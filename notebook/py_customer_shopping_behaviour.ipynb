```python
# For sake of simplicity we are inserting null values with categoty wise median value

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# will make column names in the snake case for ease fo use

df.columns = df.columns.str.lower()    # converting in lower case
df.columns = df.columns.str.replace(' ','_')    # replacing space with underscore
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels = ['Young Adult','Adult','Middle-aged', 'Senior']

df['age_group'] = pd.qcut(df['age'],q=4, labels =labels)
# 'qcut' splits ages into four equal size group based on data and assign labels 
```

```python
df[['age','age_group']].head(10)
```

```python
#create the column purchase_frequency_days 

frequency_mapping = {
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Every 3 Months': 90, 
    'Quarterly': 90,
    'Annually': 365
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# check 'discount_applied' and 'promo_code_used' columns if they have resemblance.
df[['discount_applied','promo_code_used']].head(10)
```

```python
#comparing two columns

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# they are same. will drop one of them

df = df.drop('promo_code_used', axis=1)
```

```python
df.columns
```

```python
from sqlalchemy import create_engine, Numeric

engine = create_engine('oracle://da2:da2@localhost:1521/XE')

df.to_sql(
    'customer_shopping',
    engine,
    if_exists='replace',
    index=False,
    dtype={col: Numeric(10, 2) 
           for col in df.select_dtypes(include=['float64']).columns}
)
```

```python
# export cleaned file 

df.to_csv(r'C:\Users\pramod\Documents\Python Workspace\Data\export_customer_shopping.csv', index=False)
```
